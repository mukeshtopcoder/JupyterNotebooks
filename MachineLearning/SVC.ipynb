```python
import pandas as pd
df = pd.read_csv("c:/users/saini/Downloads/500hits.csv",encoding="latin-1")
df.head(3)
```

```python
df = df.drop(columns=['PLAYER','CS'])
```

```python
df.head(3)
```

```python
X = df.iloc[:,0:13]
y = df.iloc[:,13]
```

```python
X.head(3)
```

```python
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=17,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
svc.score(X_test,y_test)
```

```python
svc_reg0 = SVC(C=0.1)
```

```python
svc_reg0.fit(X_train,y_train)
```

```python
svc_reg0.score(X_test,y_test)
```

```python
svc_reg1 = SVC(C=1)
```

```python
svc_reg1.fit(X_train,y_train)
```

```python
svc_reg1.score(X_test,y_test)
```

```python
svc_reg2 = SVC(C=1000)
svc_reg2.fit(X_train,y_train)
svc_reg2.score(X_test,y_test)
```

```python
svc_gamma0 = SVC(gamma=0.1)
svc_gamma0.fit(X_train,y_train)
svc_gamma0.score(X_test,y_test)
```
