```python
import pandas as pd
```

```python
df = pd.read_excel("c:/users/saini/Downloads/Financial_Sample.xlsx")
```

```python
df.head(3)
```

```python
df = df.drop(columns=["Month Number","Month Name","Year"])
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df = df.fillna(method="ffill")
```

```python
df.isnull().sum()
```

```python
df["Discount Band"] = df["Discount Band"].fillna("Low")
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder , OrdinalEncoder
```

```python
ohe = OneHotEncoder(sparse_output = False)
ode = OrdinalEncoder()
```

```python
from sklearn.compose import make_column_transformer
```

```python
ct = make_column_transformer(
    (ohe,["Segment"]),
    (ohe,["Country"]),
    (ohe,["Product"]),
    (ode,["Discount Band"]),
    remainder="passthrough"
)
```

```python
ct.set_output(transform="pandas")
```

```python
newDf = ct.fit_transform(df)
```

```python
newDf.shape
```

```python
newDf.head(3)
```

```python
newDf.isnull().sum().sum()
newDf.shape
```

```python
def findProfit(value):
    if(value>0):
        return 1
    else:
        return 0
```

```python
newDf["Profit/Loss"] = newDf["remainder__Profit"].map(findProfit)
```

```python
newDf.head(3)
```

```python
newDf = newDf.drop(columns=["remainder__Date"])
```

```python
newDf.shape
```

```python
y = newDf.iloc[:,25]
```

```python
y.head(3)
```

```python
X = newDf.iloc[:,0:25]
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=17,test_size=0.2)
```

```python
X_train.head(3)
```

```python
y_train.head(3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc.score(X_test,y_test)
```

```python
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
cm = confusion_matrix(y_pred,y_test)
```

```python
cm
```

```python
cr = classification_report(y_pred,y_test)
```

```python
print(cr)
```
