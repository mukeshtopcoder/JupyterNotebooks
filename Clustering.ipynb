```python
# Clusttering
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
X , y = make_blobs(n_samples=500 , centers=3 , cluster_std = 0.60 )
```

```python
len(X)
```

```python
len(y)
```

```python
X[:10]
```

```python
y[:10]
```

```python
n = len(np.unique(y))
print('Number of Culster Should be :  ',n)
```

```python
kmeans = KMeans(n_clusters = n)
```

```python
kmeans.fit(X)
```

```python
y_pred = kmeans.predict(X)
```

```python
plt.figure(figsize=(4,2))
plt.scatter(X[:,0] , X[:,1] , c=y_pred , s=10 , cmap='viridis')
plt.show()
```

```python
centers = kmeans.cluster_centers_
```

```python
centers
```

```python
plt.figure(figsize=(4,2))
plt.scatter(centers[:,0] , centers[:,1] , c=[0,1,2] , s=70 , cmap='viridis')
plt.show()
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
print('Data Points  : ',len(X))
print(X[:10])
```

```python
print('Data Points  : ',len(y))
print(y[:10])
```

```python
n = len(np.unique(y))
print('Number of Clusters Should Be : ',n)
kmeans2 = KMeans(n_clusters = n)
```

```python
kmeans2.fit(X)
```

```python
y_pred = kmeans2.predict(X)
```

```python
labels = kmeans2.labels_
```

```python
labels
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(X)
```

```python
x_pca[:10]
```

```python
plt.figure(figsize=(10,2))
sns.scatterplot(x=x_pca[:,0] , y=x_pca[:,1] , hue = y , palette='Set2')
plt.show()
```

```python
plt.figure(figsize=(10,2))
sns.scatterplot(x=x_pca[:,0] , y=x_pca[:,1] , hue = y_pred , palette='Set2')
plt.show()
```
