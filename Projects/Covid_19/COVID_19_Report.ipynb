```python
df = pd.read_csv('country_wise_latest.csv')
df.head()
```

what I have undersatnd from this data:

confirmed = death + recovered + active
Active --> the number of patient currenly in each country

WHO Region --> refers to the World Health Organization regional classification

Deaths / 100 cases--> the case fatality rate (CFR) (how deadly the virus is in each country) expressed per 100 cases.

Recovered / 100 cases --> Shows how many people recovered out of every 100 who were infected. (how well each country recovers infected patients)

Deaths / 100 recovered --> Shows how deaths relate to recoveries it's a measure of outcome severity. (high Deaths / 100 Recovered = more deaths per recovery, showing poor medical outcome)

```python
#the number of countries in each region
regions = df['WHO Region'].value_counts().reset_index()
fig = px.bar(regions, x='WHO Region' , y='count', labels={'WHO Region':'World Health Organization Regions'},title='Number of Countries in Each Region')
fig.show()
```

```python
# country that has the largest death number
df[['Country/Region','Deaths']].groupby(['Country/Region']).max().sort_values(by='Deaths',ascending=False).head(1)
```

```python
# check the country which has the highest recovered people number
df.loc[df['Recovered'] == df['Recovered'].max(), 'Country/Region']
```

```python
# the countries that has the lowest death number wich is 0
# we have seen(in the describe function) the minimum values of column death which is zero
df.loc[df['Deaths'] == 0, 'Country/Region']
```

```python
# the country which has the lowest recovered people number
# we have seen(in the describe function) the minimum values of column recovered which is zero
df.loc[df['Recovered'] == df['Recovered'].min(), ['Country/Region','Recovered']]
```

```python
# Top 5 countries that have the highest active cases now
contries = df.groupby('Country/Region')['Active'].sum().sort_values( ascending = False).head().reset_index()

fig = px.bar(contries, x = 'Country/Region',y='Active', title='Top 5 Countries that Have the Highest Active Cases', labels={'Country/Region':'Countries'})
fig.show()
```

```python
# the world map to represent the intensity of Confirmed cases in each country
fig = px.choropleth(df, locations='Country/Region',color='Confirmed',hover_name='Country/Region',locationmode="country names", color_continuous_scale=px.colors.sequential.Sunsetdark, title='Confirmed Cases In Each Countries')
fig.show()
```

We can see the distribution of the confirmed cases in each country and how America, Brazil, India, and Russia have the most cases between active, recovered, and deaths

```python
# the death rate in precentage in each region
continent_summary = df.groupby('WHO Region')[['Deaths', 'Confirmed']].sum()
continent_summary['Death Rate %'] = (continent_summary['Deaths'] / continent_summary['Confirmed']) * 100

fig = px.bar(continent_summary ,y='Deaths',labels={'WHO Region':'World Health Organization Regions'} ,title='World Health Organization Regions that Have the Higher Death Rate')
fig.show()
```

We can see that America and Europe have the highest death rates

```python
# the active rate in precentage in each region
continent_summary = df.groupby('WHO Region')[['Active', 'Confirmed']].sum()
continent_summary['Death Rate %'] = (continent_summary['Active'] / continent_summary['Confirmed']) * 100

fig = px.bar(continent_summary ,y='Active', labels={'WHO Region':'World Health Organization Regions'} ,title='World Health Organization Regions That Have The Higher Active Rate')
fig.show()
```

We can see that America and Europe have the highest active cases, and Africa has more active cases than the Eastern Mediterranean, whereas in the death rate, the Eastern Mediterranean has the highest death rate

```python
# top 10 countries that have the higher case fatality rate
higher_case_fatality = df.sort_values(by=['Deaths / 100 Cases'],ascending=False).head(10)
fig = px.bar(higher_case_fatality, x='Country/Region',y='Deaths / 100 Cases',color="WHO Region",title="Deaths per 100 COVID-19 Cases by Country",labels={"Deaths / 100 Cases": "Deaths per 100 Cases", 'Country/Region':'Countries'} ,hover_data=['Deaths','Active'])
fig.show()
```

The top 10 countries in the case fatality rate (CFR) per 100 cases across regions indicates how deadly the virus is in each country. We can see that Europe has 7 countries in this list, which can reflect a healthcare system stress in these countries, thus the high CFR rate.

```python
# top 10 countries that have the higher recovery efficiency
recovery_efficiency = df.sort_values(by=['Recovered / 100 Cases'],ascending=False).head(10)
fig = px.bar(recovery_efficiency, x='Country/Region',y='Recovered / 100 Cases',color="WHO Region" ,hover_data=['Recovered','Active'],title="Recovered per 100 COVID-19 Cases by Country",
    labels={"Recovered / 100 Cases": "Recovered per 100 Cases",'Country/Region':'Countries'})
fig.show()
```

The top 10 countries in recovery efficiency, which indicates how well each country recovers infected patients across regions. We can see that there is no specific region that has a significantly higher recovery rate by country, but the Western Pacific region has 3 countries.

```python
continent_summary = df.groupby('WHO Region')[['Deaths', 'Recovered','Active','Confirmed']].sum().reset_index()

# Melting to long format
df_long = continent_summary.melt(id_vars='WHO Region', value_vars=['Recovered','Deaths','Active','Confirmed'],var_name='Status', value_name='Count')

fig = px.bar(df_long,
             x='WHO Region', y='Count', color='Status',
             barmode='group',  # side by side bars
             title='COVID-19 Cases by WHO Region and Status',
             labels={'Count': 'Number of Cases', 'WHO Region': 'WHO Region'})
fig.update_yaxes(type="log")
fig.show()
```

comparing the different regions according to the active, recovered, death, and confirmed cases. We can tell that the virus is widely spread in the Americas region, hence the high numbers in all the status, unlike the Western Pacific region

```python
# countries with high new Deaths and low new recoveries
fig = px.scatter(df, x="New recovered", y="New deaths", color="WHO Region", hover_name="Country/Region", size="Confirmed", title="Countries with High New Deaths and Low New Recoveries")
fig.show()
```

The countries (like Peru, Mexico) in the upper left position have high new deaths and low new recoveries, which is a warning sign that the situation in such countries is worsening, the upper left position have high new deaths and low new recoveries, which is a warning sign that the situation in such countries is worsening and may be the healthcare system is falling.

```python
# focusing on the coutries with critical conditions

avg_new_death = df['New deaths'].mean()
avg_new_recovered = df['New recovered'].mean()

critical = df[(df['New deaths'] > avg_new_death) & (df['New recovered'] < avg_new_recovered)]
px.bar(critical, x='Country/Region', y='New deaths', color='WHO Region', labels={'Country/Region':'Countries'})
```

Those two countries are considered critical conditions because they have above-average new deaths, and the new recoveries are below average in each of them are in the Americas region.

```python
# High 1 week % increase (speed of the outbreak) with high Active (size of the outbreak)â†’ outbreak growing
# Finds fast-growing and heavily affected countries
avg_increase = df['1 week % increase'].mean()
avg_active = df['Active'].mean()
heavily_affected_countries = df[(df['1 week % increase'] > avg_increase) & (df['Active'] > avg_active)]

fig = px.scatter(heavily_affected_countries, x='1 week % increase',
y='Active', size='Confirmed', color='WHO Region', hover_name='Country/Region', title='COVID-19 Hotspots: High Growth and High Active Cases')
fig.show()
```

Identifying heavily affected countries where both 1-week percentage increases and the active cases are above average, those countries are identified as hotspots with fast-growing outbreaks, and many people are currently infected across the region. We can see that India has the highest increase in one week, with the highest active cases.
