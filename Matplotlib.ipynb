```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
dic = {
    'City':['Noida','Delhi','Moradabad','Gaziabad','faridabad','Delhi','Noida'],
    'Sales':[23456,86763,56433,41235,56788,34565,24556]
}
```

```python
df = pd.DataFrame(dic)
```

```python
df
```

```python
plt.bar(df['City'],df['Sales'])
plt.title('Sales By City')
plt.show()
```

```python
sbc = df.groupby(['City']).agg({'Sales':'sum'})
```

```python
sbc = sbc.reset_index()
```

```python
sbc
```

```python
plt.bar(sbc['City'],sbc['Sales'])
plt.title('Sales By City')
plt.show()
```

```python
cbp = (df.groupby(['Category']).agg({'Purchase Amount (USD)':'sum'})).reset_index()
```

```python
plt.plot(cbp['Category'],cbp['Purchase Amount (USD)'],marker='o',mfc='r',mec='r',ms=10,ls=':',color='r')
plt.title('Purchase By Category')
plt.xlabel('Category')
plt.ylabel('Purchase Amount')
plt.grid()
plt.show()
```

```python
plt.plot(cbp['Category'],cbp['Purchase Amount (USD)'],marker='o',mfc='r',mec='r',ms=10,ls=':',color='r')
plt.title('Purchase By Category')
plt.xlabel('Category')
plt.ylabel('Purchase Amount')
plt.grid(axis='y')
plt.show()
```

```python
plt.plot(cbp['Category'],cbp['Purchase Amount (USD)'],marker='o',mfc='r',mec='r',ms=10,ls=':',color='r')
plt.title('Purchase By Category')
plt.xlabel('Category')
plt.ylabel('Purchase Amount')
plt.grid(color='b' ,linestyle='dotted',linewidth=2)
plt.show()
```
