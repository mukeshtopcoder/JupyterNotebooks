# DBSCAN -> Density Based Spatial Clustering of Applications with Noise
- Dosen't Need to specify the number of clusters upfront.
- Can find arbitrarily shaped clusters.
- Can identify outliers as noise.
- Great when clusters are of different shapes or densities.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
X , y = make_moons(n_samples=500 , noise=0.05 )
```

```python
X.shape
```

```python
y.shape
```

```python
X = StandardScaler().fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.3 , min_samples=5)
```

```python
labels = dbscan.fit_predict(X)
```

```python
plt.figure(figsize=(6,3))
plt.scatter(X[:,0],X[:,1],c=labels , cmap="plasma",s=30)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
unique_labels = set(labels)
```

```python
unique_labels
```

```python
for label in unique_labels:
    label_mask = labels == label
    plt.scatter(X[label_mask,0],X[label_mask,1],label=f'Cluster {label}' if label != -1 else 'Noise')
```

# Agglomerative Clustering
- It's a Bottom-Up Approach.
  - Each point starts in its own cluster
  - Clusters are **merged step by step** based on similarity(distance).
- You don't need to predefine shapes of clusters like in K-Means.
- Often visualized using a dendrogram (tree structure).

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
```

```python
# For generating Synthetic data
# centers = no of cluster
# cluster_std = Cluster Standard Divition Distance
X , y = make_blobs(n_samples=500 , centers=4 , cluster_std = 0.60 )
```

```python
X.shape
```

```python
y.shape
```

```python
agg = AgglomerativeClustering(n_clusters=4)
```

```python
labels = agg.fit_predict(X)
```

```python
labels
```

```python
plt.figure(figsize=(6,3))
plt.scatter(X[:,0],X[:,1],c=labels , cmap="viridis",s=30)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram , linkage
```

```python
linked = linkage(X , method="complete")
```

```python
plt.figure(figsize=(3,3))
dendrogram(linked)
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```
